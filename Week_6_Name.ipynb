# 2. Binding

**Binding** = association between two things.

```
name → variable
variable → type
variable → memory location
function → code
```

### Example

```python

```

Bindings created:

```
x → object
x → type(int)
```

## 2.1 Binding Issues

When **associations become unclear or incorrect**. Some common ones:

```
multiple names → same object (alias)
one name → multiple objects (polymorphism)
binding changes over time
```

Examples: Alias problem




````python
# This was made in 2022

print('Before the match:')
ricon = ['gáy']
fan_mp3 = ricon
print('fan_mp3[0] = ricon[0] =', fan_mp3[0]) # -> gáy

print('\nAfter the match: fan_mp3[0] changed to "cay"')
fan_mp3[0] = 'cay'
print('ricon[0]:', ricon[0]) # -> 'cay'
````

- `fan_mp3` and `ricon` point to the same thing.
- Modifying `fan_mp3` also changes `ricon`.

# 3. Object Lifetime
- Object: any program entity
- **Lifetime:** Period of time for
`
creation → destruction
`

Proper example:
```c
p = new int
delete p
```

Problems occur when lifetime is mismanaged.

### 3.1 Dangling Reference

A **dangling reference** occurs when a pointer references an object that has already been destroyed.

```cpp
int* p = new int;
int* q = p;
delete p;

*q = 5   // invalid access
```

Memory situation: `
q → memory that no longer exists
`


### 3.2 Memory Leak

A **memory leak** occurs when memory is allocated but no reference remains to free it.

```cpp
int* j97_fatherhood = new int;
j97_fatherhood = NULL;
```
- allocated memory cannot be accessed
- memory permanently lost

# 5. Scope
Textual region where a binding is effective/visible.

### 5.1 Static Scope (Lexical Scope)

**Definition**

- Binding of a variable is determined by the **program structure (the blocks where code is written)**.
- Compiler can determine this **before the program runs**.

```
Static scope follows program structure
(block nesting)
```

### 5.2 Dynamic Scope

- Binding of a variable is determined by the **runtime call sequence**.
- The variable used is the **most recently created active declaration**.


```
Dynamic scope follows the call stack
(runtime execution order)
```

### 5.3 Example

```python
x = 10

def f():
    print(x)

def g():
    x = 20    # Where the difference happens
    f()

g()
```



- Static scope: prints `10` (priority to global `x`)
- Dynamic scope: prints `20` (most recent declaration)


| Static Scope                        | Dynamic Scope                            |
| ----------------------------------- | ---------------------------------------- |
| Determined at compile time          | Determined at runtime                    |
| Follows **program structure**       | Follows **call stack**                   |
| Lookup through **enclosing blocks** | Lookup through **active function calls** |
| Used by most modern languages       | Rare today                               |

Example languages:

```
Static: Python, C, Java, Scala
Dynamic: early Lisp, some scripting systems
```

# 7. Static Scope Rules
- Use the **most local declaration**
- Declarations invisible (no effect) outside their block
- Inner blocks can access outer variables (unless re-declared inside)
- Naming a block = local declaration in its outer block


````python
x = 'cá mập'

def at_home():
    x = 'cá con'
    print(f'Ở nhà anh là: {x}')

print(f'Ra đường anh là: {x}')
at_home()
````

# 9. Python Scope Chain

Python lookup order: LEGB rule

```
Local
↓
Enclosing
↓
Global
↓
Built-in / imports
```
Example:
```python
x = 3 # global

def f():
    y = 4 # local of f
```




Introducing: `nonlocal` and `local` in Python

````python
x = 3   # x: global
y = 4   # y: global

def f():
    x = 69    # x: f
    z = 1     # z: f
    def g():
        y = 1         # y: g
        nonlocal x    # x: f

        # 'nonlocal z' not in here: z can only be read, not modified
        x += z
    g()
    print(f'In {x} possible actions, I like sleeping, the rest is {x-z}')

def h():
    global x    # x: global
    x = 6969

f()
print(f'global x before h(): {x}')

h()
print(f'global x after h(): {x}')
````
