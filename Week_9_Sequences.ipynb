# 1. Expressions

An **expression** is a syntactic entity whose evaluation either:

* Produces a value
* Fails to terminate
* Becomes undefined

Examples:

```c
4 + 3 * 2
(a + b) * (c - a)
(b != 0) ? (a / b) : 0
```

Key idea:

> Expressions compute values.

---

## 1.1 Expression Evaluation

Expressions have a **functional composition** structure.

Example:

```c
(a + b) * (c - d)
```

Tree form:

```text
        *
      /   \
     +     -
    / \   / \
   a   b c   d
```

Evaluation usually happens by evaluating subexpressions first, then applying the parent operator.

---

## 1.2 Expression Syntax

Expressions may be written in different forms:

| Notation | Example             |
| -------- | ------------------- |
| Infix    | `(a + b) * (c - d)` |
| Prefix   | `* + a b - c d`     |
| Postfix  | `a b + c d - *`     |

Same meaning, different syntax.

### 1.2.1 Infix Notation

Infix notation places the operator between operands.

Example:

```c
(a + b) * (c - d)
```

Common in imperative languages:

* C
* Java
* Python
* Pascal

Good for binary operators.

Problem:

> Without precedence and associativity, infix expressions can be ambiguous.


### 1.2.2 Prefix Notation

<img src="https://raw.githubusercontent.com/tuankhoin/CO3057-Computer-Vision/refs/heads/main/assets/m_mixup.JPG" height=300 />

Prefix notation places the operator before operands.

Example:

```text
* + a b - c d
```

Equivalent to:

```c
(a + b) * (c - d)
```

Other forms:

```text
(* (+ a b) (- c d))
*(+(a,b),-(c,d))
```

Advantage:

> No precedence rules needed if operator arity is known.

Common in:

* Lisp
* Function-style languages
* AST notation

### 1.2.3 Postfix Notation

Postfix notation places the operator after operands.

Example:

```text
a b + c d - *
```

Equivalent to:

```c
(a + b) * (c - d)
```

Common uses:

* Factorial: `5!`
* Stack machines
* Intermediate code
* PostScript

Important compiler idea:

> Postfix is easy to evaluate using a stack.

---


## 1.3 Precedence

Precedence decides **which operator binds first**.

Example:

```c
3 + 4 * 5
```

Result:

```text
3 + (4 * 5) = 23
```

Not:

```text
(3 + 4) * 5 = 35
```

Languages usually follow mathematical convention, but exact precedence levels may differ between languages.

---

## 1.4 Associativity

Associativity is used when operators have the **same precedence**.

Example:

```c
a - b - c
```

Usually interpreted as:

```c
(a - b) - c
```

This is **left associative**.

Some operators are right associative:

```c
a = b = c
```

interpreted as:

```c
a = (b = c)
```

---

## 1.5 Why Associativity Matters

In mathematics, some operators are associative:

```text
(a + b) + c = a + (b + c)
```

But in computers, this may fail because of:

* Overflow
* Floating-point rounding
* Side effects

Example:

```text
10^20 * 10^20 * 10^-20
```

Changing grouping can change intermediate values.

---

## 1.6 Parentheses

Parentheses override precedence and associativity.

Example:

```c
(A + B) * C
```

Parentheses make the intended grouping explicit.

A language could theoretically remove precedence rules and require parentheses everywhere.

Tradeoff:

| Advantage    | Disadvantage         |
| ------------ | -------------------- |
| Simple rules | Harder to write/read |

## 1.7 Conditional Expressions

A conditional expression chooses a value based on a condition.

Statement form:

```c
if (count == 0)
    average = 0;
else
    average = sum / count;
```

Expression form:

```c
average = (count == 0) ? 0 : sum / count;
```

Key idea:

> A conditional expression returns a value.

---



## 1.8 Operand Evaluation Order

Question:

```c
int a = 5;

int fun1() {
    a = 17;
    return 3;
}

void main() {
    a = a + fun1();
}
```

What is the final value of `a`?

Possible answers:

```text
8 or 20
```

Reason:

> The result depends on whether `a` or `fun1()` is evaluated first.

This is caused by **side effects**.

### Side Effects

A side effect occurs when evaluating an expression changes program state.

Example:

```c
a = a + fun1();
```

`fun1()` changes `a`.

Expressions with side effects can make evaluation order matter.

# 2. Evaluation Mechanisms

## 2.1 Eager Evaluation

Evaluate all operands first, then apply the operator.

Example:

```c
f(a(), b())
```

`a()` and `b()` are evaluated before `f`.

---

## 2.2 Lazy Evaluation

- Operands are passed unevaluated.
- The operator/function decides whether they are needed.

Example:

```c
a == 0 ? b : b / a
```

If `a == 0`, then `b / a` should not be evaluated.

Most languages use:

* Lazy evaluation for conditionals / boolean short-circuit
* Eager evaluation for ordinary expressions

---

## 2.3 Short-Circuit Evaluation

Short-circuit evaluation skips unnecessary operands.

Example:

```c
(a == 0) || (b / a > 2)
```

If `a == 0` is true,
second operand is skipped ⟶ This prevents division by zero.

---

## 2.4 Short-Circuit and Side Effects

Example:

```c
(a > b) || (b++ / 3)
```

If `a > b` is true:

```text
b++ is not executed
```

So short-circuit evaluation can affect program state.

Some languages provide both:

| Short-circuit              | Non-short-circuit |
| -------------------------- | ----------------- |
| Ada: `and then`, `or else` | Ada: `and`, `or`  |

# 3. Statements

A **statement** is a syntactic entity whose evaluation:

* Does not return a value
* Changes the state of the system

Examples:

```c
a = 5;
print("pippo");
begin ... end
```

Key idea:

> Expressions compute values; statements perform actions.

---

## 3.1 Assignment Statements

General form:

```text
leftExpr assignOperator rightExpr
```

Example:

```pascal
a := b;
```

Meaning:

* Evaluate right side to get a value
* Evaluate left side to get a location
* Store value into that location

Important ambiguity:

> Should the left or right side be evaluated first?

Some languages specify this; some leave it to implementation.

---

## 3.2 Assignment as Expression

C-based languages treat assignment as an expression.

Example:

```c
while ((ch = getchar()) != EOF) {
    ...
}
```

Here:

```c
ch = getchar()
```

both:

* Assigns a value to `ch`
* Returns the assigned value

Useful, but can cause side effects.

---

## 3.3 Compound and Unary Assignment

Examples:

```c
a += 1;
a -= 2;
i++;
--j;
```

Benefits:

* Shorter code
* Often clearer for simple updates

Risk:

> Can hide side effects inside expressions.

Example: `x = i++ + ++i;`

# 4. Control Structures (if-else, switch, loops,...)

Control structures determine **which statements execute and how many times**.

Two main categories:

1. Selection: Choose among alternatives
2. Iteration: Repeat statements

A control structure includes:

* The control statement
* The controlled statement(s)

---

# 5. Two-Way Selection

<img src="https://raw.githubusercontent.com/tuankhoin/CO3057-Computer-Vision/refs/heads/main/assets/m_condition.JPG" height=300 />

General form:

```text
if control_expression
then clause
else clause
```

Example:

```c
if (x > 0)
    y = 1;
else
    y = -1;
```

This is fundamental in almost all programming languages.

---

## 5.1 Dangling Else

Problem:

```c
if (sum == 0)
    if (count == 0)
        result = 0;
else
    result = 1;
```

Question:

> Which `if` does the `else` belong to?

Most languages bind `else` to the nearest unmatched `if`.

<img src="https://github.com/tuankhoin/CO3005-PPL/blob/main/assets/dllt_disapprove.png?raw=true" alt="drawing" width="200"/>

Better style:

```c
if (sum == 0) {
    if (count == 0) {
        result = 0;
    }
} else {
    result = 1;
}
```

## 5.2 Avoiding Dangling Else

Languages solve this differently:

| Language style       | Solution                |
| -------------------- | ----------------------- |
| C / Java             | Use braces `{}`         |
| Python               | Indentation             |
| Ada / Ruby / Fortran | Explicit ending keyword |

Example in Python:

```python
if sum == 0:
    if count == 0:
        result = 0
else:
    result = 1
```

Indentation determines structure.

# 6. Multiple Selection

Multiple selection chooses one among many alternatives.

Examples:

```c
switch (x) { ... }
```

Design questions:

* What type can selector have?
* How are cases specified?
* Execute one segment or multiple?
* Are ranges allowed?
* What happens if no case matches?

## 6.1 Multiple Selection: C

Example:

```c
switch (index) {
case 1:
case 3:
    odd += 1;
    sumodd += index;
    break;

case 2:
case 4:
    even += 1;
    sumeven += index;
    break;

default:
    printf("Error in switch");
}
```

Properties:

* Selector type: integer-like
* Case labels: exact values
* `break` prevents fall-through
* `default` handles unmatched values

## 6.2 C Switch: Fall-Through

In C:

```c
case 1:
case 3:
    odd += 1;
```

If `index` is 1, execution continues into case 3’s statements.

This is called **fall-through**.

Use `break` to stop execution.

Important:

> Missing `break` is a common bug.

## 6.3 Multiple Selection: Pascal

Example:

```pascal
case exp of
    1: clause_A
    2, 7: clause_B
    3..5: clause_C
    10: clause_D
else
    clause_E
end
```

Properties:

* Selector: integer or character
* Cases can be:

  * Single values
  * Multiple values
  * Subranges
* `else` handles unmatched values

# 7. Iterative Statements (Loops)

Iterative statements repeat a statement or block.

They may execute:

* Zero times
* One time
* Many times

Without loops, programs would be:

* Huge
* Inflexible
* Hard to maintain

## 7.1 Loop Design Questions

Important design questions:

1. How is iteration controlled?

   * Counter
   * Boolean condition
   * Data structure

2. Where is the condition checked?

   * Before loop body
   * After loop body
   * User-located exit

## 7.2 Counter-Controlled Loops

A counter-controlled loop has:

* Loop variable
* Initial value
* Terminal value
* Step size

Example:

```pascal
for i := first to last by step do
    body
end
```

Used when the number of iterations is known or easily determined.

## 7.3 Algol-Style Counter Loop

General form:

```pascal
for i := first to last by step do
    loop body
end
```

Semantics:

```text
end_save := last
i := first

loop:
    if i > end_save goto out
    loop body
    i := i + step
    goto loop

out:
```

Important detail:

> The terminal value is saved before the loop begins.

This makes the number of iterations more predictable.

## 7.4 C-Style For Loop

General form:

```c
for (expr1; expr2; expr3)
    body
```

Semantics:

```text
expr1

loop:
    if expr2 == 0 goto out
    body
    expr3
    goto loop

out:
```

C `for` is very flexible.

Example infinite loop:

```c
for (;;) {
    ...
}
```

## 7.5 Logically Controlled Loops

A logically controlled loop repeats based on a Boolean expression.

Examples:

```c
while (condition) {
    ...
}
```

```c
do {
    ...
} while (condition);
```

More general than counter-controlled loops.

---


# 8. Pretest vs Posttest

## 8.1 Pretest loop

Condition checked before body.

```c
while (condition) {
    body
}
```

May execute zero times.

## 8.2 Posttest loop

Condition checked after body.

```c
do {
    body
} while (condition);
```

Executes at least once.

# 9. C Loop

## 9.1 C While Loop Semantics

```c
while (ctrl_expr)
    body
```

Equivalent control flow:

```text
loop:
    if ctrl_expr is false goto out
    body
    goto loop

out:
```

## 9.2 C Do-While Loop Semantics

```c
do
    body
while (ctrl_expr);
```

Equivalent control flow:

```text
loop:
    body
    if ctrl_expr is true goto loop
```

Key difference:

> `do-while` always executes the body at least once.

# 10. User-Located Loop Control

Sometimes the programmer wants to exit or continue in the middle of a loop.

Common statements:

```c
break;
continue;
```

These are restricted forms of `goto`.

---

## 10.1 Break

Example:

```c
while (sum < 1000) {
    getnext(value);
    if (value < 0) break;
    sum += value;
}
```

Meaning:

```text
If value < 0, exit the loop immediately.
```

---

## 10.2 Continue

If `break` is replaced by `continue`:

```c
while (sum < 1000) {
    getnext(value);
    if (value < 0) continue;
    sum += value;
}
```

Meaning:

```text
If value < 0, skip sum += value and continue next iteration.
```

Important difference:

| Statement  | Effect                  |
| ---------- | ----------------------- |
| `break`    | exits loop              |
| `continue` | skips to next iteration |

---

# 11. Iteration Based on Data Structures

Loop controlled by elements in a data structure.

Example idea:

```text
for each element in collection
```

Uses an **iterator**.

Iterator:

* Called at each iteration
* Returns the next element
* Has a defined traversal order
* May be predefined or user-defined

## 11.1 Java Iterator Example

```java
ArrayList al = new ArrayList();
Iterator it = al.iterator();

while (it.hasNext()) {
    Object element = it.next();
    ...
}
```

Meaning:

* `hasNext()` checks if more elements exist
* `next()` returns the next element

## 11.2 C# Foreach Example

```csharp
String[] strList = {"Bob", "Carol", "Ted"};

foreach (String name in strList)
    Console.WriteLine("Name:{0}", name);
```

Meaning:

> Loop once for each element in the array.

This hides iterator details from the programmer.

# 12. Unconditional Branching

An unconditional branch jumps directly to another statement.

Classic form:

```c
goto label;
```

Powerful, but dangerous.

Problems:

* Hard to read
* Hard to maintain
* Easy to create unreliable code

## 12.1 Goto and Structured Programming

Structured programming discourages unrestricted `goto`.

Modern languages often remove it:

* Java: no general `goto`
* Python: no `goto`
* Ruby: no `goto`

But restricted forms remain:

```text
break
continue
return
```

These are controlled jumps.
