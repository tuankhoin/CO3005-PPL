# 1. Data Types

## Definition

A data type consists of:

* A set of values
* A set of operations on those values

**Purpose:**

* Organize programs conceptually
* Detect errors
* Support implementation

# 2. Type System

A type system includes:

* Predefined types
* Mechanisms to define new types
* Rules for:

  * Type equivalence
  * Type compatibility
  * Type inference
* When type checking happens:

  * Static (compile time)
  * Dynamic (runtime)

**Key idea:**
A type system prevents illegal operations.

# 4.3 IEEE-754 (Concept)

$$Number = (-1)^{Sign} × (1+Fraction) × 2^{Exponent − Bias}$$

A floating-point number consists of:

* Sign
* Exponent
* Mantissa (fraction)

**Common formats:**

* Float (32-bit)
* Double (64-bit)

**Example**

$$5.75 = 101.11_2 = (1+\underbrace{.0111}_{Mantissa}) × \underbrace{2^2}_{Bias\;(Power\; Number)}$$

- Positive number ⟶ Sign = 0
- Bias = 2
- Exponent = $Bias + 127 = 129 = 10000001_2$
- Mantissa/Fraction = `0111`

Play around with this [Conversion website](https://www.h-schmidt.net/FloatConverter/IEEE754.html) to know more.

# 6. User-Defined Ordinal Types

An ordinal type:

* Is finite
* You can order the set of values (i.e. what is bigger than what)

**Examples:**

* Integer
* Character
* Boolean
* Enum

**Non-ordinal examples**

* Float (not finite)
* String (cannot be ordered - you cannot say something like `'CR7' ><= 'M10'`)

---

# 7. Enumeration Types

Defines a set of named constants:

* Example: days of the week

```c
enum day {Sun, Mon, Tue, Wed, Thu, Fri, Sat};

// Mon = 2, Tue = 3,...
day tgif = Fri;
```

**Advantages:**

* Improves readability
* Restricts values
* Prevents invalid assignments

---

# 8. Subrange Types

A subset of an existing ordinal type

**Example:**

```c
type sv = array[1 .. 50] of string;
```

**Purpose:**

* Restrict values at compile time

# 16. Pointer Types `*ptr`

* Store memory addresses

**Used for:**

* Dynamic memory
* Indirect access

Operations:
* Assignment (set value to an address)
* Dereferencing (access value at address)
  * C++ Example: `j = *p` sets `j` to values stored in `p`.

Problems
* Dangling pointer (points to freed memory)
* Memory leak (memory lost, cannot access → Garbage)

---

# 17. References vs Pointers

Remember:
- `&A`: address of A
- `*A`: value at A's address

Difference:
* Pointer: stores address
* Reference: alias to a variable

| Reference Type             | Pointer            | Notes                                                                 |
|----------------------------|--------------------|----------------------------------------------------------------------|
| `int A;`                   | `int A;`           | Step 1: Both declare an integer variable.                                    |
| `int& rA = A;`             | `int* pA = &A;`    | Step 2: A reference must be initialized to an existing variable; a pointer stores the address of a variable. |
| `rA ⇒ A`                   | `*pA ⇒ A`          | Dereferencing a reference is implicit; dereferencing a pointer requires `*`. |
| `N/A`                      | `pA++`             | References cannot be incremented; pointers can move to the next memory location. |
| `cannot reseated`           | `pA = &B`          | A reference always refers to the same variable; a pointer can be reassigned to another variable’s address. |
| `cannot be null`           | `pA = null`        | References must always refer to a valid object; pointers can be null. |
| `cannot be uninitialized`  | `int* pA`          | References must be initialized at declaration; pointers can be declared without initialization. |
