## III. What is this subject all about?

- You have learned many programming languages
- They are different from each other, you might have different impressions

![](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSUWDRXEYLvQN2BEXXppTht8gCGeWLNTT3Wog&s)

![](https://i.redd.it/n2wps3a1m8n91.jpg)

![](https://external-preview.redd.it/h2sze0evX-oipC2QO-TGtmdVcji6k6Q_XNuKSlq2Ihk.jpg?auto=webp&s=ad22a598d85cf954a36cfb0eda763aeb08e0988b)

- But they all share the same thing as a programming language: Communication to the machine!

In simple term, it is:

```Human thought -> Machine Behavior```

In details, it is:

```lexer → parser → AST → name binding → typing → runtime → machine```

It all boils down to the deisgn: keyword, rule, error message, and runtime feature, etc.
