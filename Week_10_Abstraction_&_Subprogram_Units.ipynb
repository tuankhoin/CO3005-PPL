# 1. Subprogram Definition

A subprogram consists of:

### Specification

* Name
* Parameters

  * input / output
  * order
  * type
  * passing mechanism
* Behavior

### Implementation

* Local data
* Sequence of statements (body)

```text
Specification = interface
Implementation = actual code
```

---

## 1.1 Example Insight

A single piece of code may define:

* Multiple subprograms
* Nested subprograms
* Anonymous functions

Modern languages (Scala, Java) allow functions inside functions.

---

## 1.2 Subprogram Activation

A subprogram **activation**:

* Created when function is called
* Destroyed when function returns

Activation contains:
* Static: Code segments
* Dynamic:
  * Parameters
  * Local variables
  * Return address
  * Links (control/static links)

```text
Activation = runtime instance of a function call
```

Multiple activations exist during recursion.

# 2. Subprogram Mechanisms

Types of execution mechanisms:

* Simple call-return
* Recursive call
* Exception handling
* Coroutines
* Scheduled subprograms
* Tasks (concurrency)

---

## 2.1 Simple Call-Return

Characteristics:

* No recursion
* Explicit call site
* Single entry point
* Immediate execution
* Runs once, returns

Flow:

```text
caller → callee → return → caller
```

---

## 2.2 Recursive Call

A subprogram calls itself:

* Direct recursion
* Indirect (mutual) recursion

Key Requirement:

```text
Multiple activation records
```

Each call must have its own data.

---

## 2.3 Exception Handling

Used for:

* Error handling
* Event-driven programming

Example:

```java
if (length(V) == 0)
    throw new EmptyExcp();
```

Flow:

```text
throw → search handler → execute handler
```

Design Questions - A language must define:

* How exceptions are defined
* How they are raised
* How they are handled

| Language | Mechanism             |
| -------- | --------------------- |
| Java     | subclass of Throwable |
| Ada      | special type          |
| C++      | any value             |

---

## 2.4 Raising Exceptions

Exceptions can be triggered by:

* User interaction
* OS events
* Objects (timers)
* Programmer (`throw`)

---

## 2.5 Handling Exceptions

Requires:

* Protected block (`try`)
* Handler (`catch`)

### Semantics

- Non-resumable (common)
  * Stop execution
  * Unwind stack
- Resumable (rare)
  * Continue execution

# 3. Parameter Passing

Two roles:

* Formal parameters → in definition
* Actual parameters → in call

Example:

```c
int foo(float a, int b)
```

---

## 3.1 Matching

### By position

```text
foo(x+1, y-2)
```

👉 Most common (C, Java)

* Simple, fast

- Easy to mismatch arguments

### By name

```text
foo(b = x+1, a = y-2)
```

👉 Used in Python, Ada

* More readable (especially many params)

- Slightly more overhead

---

## 3.2 Parameter Passing Types

### Input-Output

* Pass by value-result
* Pass by reference
* Pass by name

👉 Used when function must **modify caller variables**

### Input Only

* Pass by value
* Pass by constant reference

👉 Used when function only **reads data**


### Output Only

* Pass by result
* Return value

👉 Used when function produces results

---

## 3.3 Pass by Value

```text
copy value → function
```

Changes inside function do NOT affect caller.

👉 When to use:

* primitives (int, float)
* avoid side effects

**Example:**

```c
void f(int x) {
    x = 10;
}

int a = 5;
f(a);
// a is still 5
```

**Pros:**

* Safe, no aliasing
* Easy to reason

**Cons:**

* Copy overhead (large data)
* Cannot modify caller

---

## 3.4 Pass by Reference

```text
pass address → function
```

Changes affect caller directly.

👉 When to use:

* need to modify caller
* large objects (avoid copy)

**Example:**

```cpp
void f(int &x) {
    x = 10;
}

int a = 5;
f(a);
// a becomes 10
```

**Pros:**

* Efficient
* Supports multiple outputs

**Cons:**

* Side effects
* Aliasing bugs

---

## 3.5 Pass by Value-Result

Modify under a different name.

<img src="https://raw.githubusercontent.com/tuankhoin/CO3057-Computer-Vision/refs/heads/main/assets/m_transfer.JPG" height=300 />

```text
copy in → modify → copy out
```

👉 When to use:

* want modification but avoid direct aliasing during execution

**Example (conceptual):**

```c
void f(int x) {   // behaves like value-result
    x = x + 1;
}

int a = 5;
f(a);
// a becomes 6 after return
```

**Aliasing issue:**

```c
f(a, a); // which value is written back?
```

**Pros:**

* Avoids aliasing during execution

**Cons:**

* Copy twice (slow)
* Undefined with same variable passed multiple times

---

## 3.6 Pass by Name

Conceptually:

```text
parameter = expression substituted into function
```

Equivalent to **macro-like substitution**

👉 When to use:

* mostly theoretical (Algol)
* delayed evaluation

**Example intuition:**

```c
f(a[i])
```

Inside `f`, every use re-evaluates `a[i]`

**Pros:**

* Very flexible
* Re-evaluates expression each time

**Cons:**

* Hard to reason
* Expensive
* Side effects multiply

---

## 3.7 Key Difference Summary

| Method       | Behavior               | Key Use               |
| ------------ | ---------------------- | --------------------- |
| Value        | copy                   | safety                |
| Reference    | alias                  | efficiency + mutation |
| Value-result | copy-in/out            | controlled mutation   |
| Name         | re-evaluate expression | flexibility           |

---

## 3.8 Input Only Parameters

### Pass by value

* Safe
* Copy overhead

👉 Preferred for:

* primitives
* simple functions

---

### Pass by Constant reference


<img src="https://raw.githubusercontent.com/tuankhoin/CO3057-Computer-Vision/refs/heads/main/assets/m_assign.JPG" height=300 />

```cpp
const int& x
```

👉 When to use:

* large objects (vector, struct)

**Example:**

```cpp
void print(const vector<int>& v) {
    // cannot modify v, only access its values
}
```

**Pros:**

* No copy
* Safe

---

## 3.9 Output Only Parameters

### Pass by result

* Only output matters
* Value assigned on return

👉 Rare in modern languages

---

### Function return

```c
return value;
```

👉 Preferred modern style

**Example:**

```c
int sum(int a, int b) {
    return a + b;
}
```

---

## 3.10 Practical Notes

### Java / Python behavior

```text
pass-by-value (of reference)
```

Example:

```python
def f(x):
    x.append(10)

a = [1,2]
f(a)
# a becomes [1,2,10]
```

👉 Reference is copied, object is shared

---

### Common mistakes

```c
x = i++ + ++i;
```

👉 Multiple side effects → undefined / confusing

---

### Key takeaway

```text
Choose based on:
- safety (value)
- efficiency (reference)
- need to modify (reference / value-result)
```

# 4. Higher-Order Functions

A function is **higher-order** if it:

* Takes functions as input
* Returns functions

Example in C:

```c
qsort(array, size, comparator)
```

`comparator` is a function passed as parameter.

👉 Common use cases:

* Sorting (custom comparator)
* Callbacks (event handling)
* Functional programming (map, filter)

---

## 4.1 Functions as Parameters

Key concept:

```text
function + environment
```

👉 A function is not just code — it may depend on surrounding variables.

---

### Binding Problem

When passing functions:

* Which variable values should be used?

👉 This matters when the function **uses external variables**.

---

### Deep Binding

Use environment at **definition time**

👉 Function remembers where it was created

---

### Shallow Binding

Use environment at **call time**

👉 Function uses current environment when called

---

### Example

```c
int x = 10; // Deep binding comes here

void f() {
    printf("%d\n", x);
}

void g(void (*func)()) {
    int x = 20; // Shallow binding comes here
    func();
}

g(f);
```

👉 Result depends on binding:

* Deep binding → prints **10**
* Shallow binding → prints **20**

---

### Example Insight

Different combinations of:

* Static vs dynamic scope
* Deep vs shallow binding

→ produce different results

---

### When this matters

* Callbacks
* Passing functions across scopes
* Functional programming

---

## 4.2 Functions as Return Values

Functions can return:

* Code
* Environment

Example:

```c
function F() {
    int x = 1;
    function g() {
        return x + 1;
    }
    return g;
}
```

---

### Problem

When `g` is returned:

👉 Where is `x`?

---

### Solution

Use **closure**: use last reference from outer environment/code block

```text
function + environment
```

---

### Example

```python
def f():
    x = 1   # Last referenced
    def g():
        return x + 1
    return g

h = f()
print(h())   # 2
```

👉 `g` still remembers `x`

---

### Use cases

* Function factories
* Delayed computation
* Encapsulation of state

---

## 4.3 Closure Concept

A closure contains:

* Code
* Captured variables

---

### Without closure

```text
function loses access to x
```

👉 leads to undefined / error

---

### With closure

Function carries its environment forward:

```text
g = (code of g) + (x = 1)
```

---

### Practical Example

```python
def make_counter():
    count = 0     # Will not be destroyed when function returns
    def inc():
        nonlocal count
        count += 1
        return count
    return inc

c = make_counter()
print(c())  # 1
print(c())  # 2
```

👉 `count` persists across calls

---

### Pros

* Enables powerful abstractions
* Keeps state without global variables
* Core of functional programming

---

### Cons

* Harder to understand/debug
* Hidden state
* Memory overhead (captured variables)

---

### Key Insight

```text
Functions are not just code. They carry context
```

---

### Final takeaway

```text
Higher-order functions = functions + behavior + context
```
