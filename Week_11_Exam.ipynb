## 1. IEEE-754 Floating Point Conversion

### Core Ideas

* Floating-point numbers are stored using: Sign bit, Exponent, Fraction (mantissa)
* IEEE-754 single precision: 1 sign bit, 8 exponent bits, 23 fraction bits
  * Values change depending on the type. Notice this carefully
* Value formula:

```text
(-1)^sign × 1.fraction × 2^(exponent-bias)
```

### Quick Example

**Question 1:**
A floating-point type follows IEEE-754 style and uses 9 bits: 1 sign bit, 4 exponent bits, and 4 fraction bits. Represent the decimal number `6.50`.

**Answer:**
- `6.5 = 110.1₂ = 1.101 × 2²`
- Bias = `2^(4-1)-1 = 7`, so exponent = `2 + 7 = 9 = 1001₂`
- Sign = `0`, fraction = `1010`
- Final: `0 1001 1010`

**Question 2:**
A floating-point type follows IEEE-754 style and uses 16 bits: 1 sign bit, 6 exponent bits, and 9 fraction bits. Again, represent the decimal number `6.50`.

**Answer:**
- `6.5 = 110.1₂ = 1.101 × 2²`
- Bias = `2^(6-1)-1 = 31`, so exponent = `2 + 31 = 33 = 100001₂`
- Sign = `0`, fraction = `101000000`
- Final: `0 100001 101000000`

### Practice Question

Which IEEE-754 component determines whether a number is positive or negative?

> Answer: Sign bit
> * 0 → positive
> * 1 → negative

# 4. Type Inference & Determining Function Types

Example: `(int, int) -> int`: function taking 2 ints and returning int


### Reading Nested Function Types

Example:

```text
(int * int * int * (int -> bool)) -> int
```

Read from inside out:

1. `(int -> bool)`: a function taking int and returning bool

2. Entire parameter list: `(int, int, int, (int -> bool))`
  * parameter 1 = int
  * parameter 2 = int
  * parameter 3 = int
  * parameter 4 = function `(int -> bool)`

3. Final return type: `-> int`

So the whole thing means:

```text
A function taking:
- 3 integers
- 1 function (int -> bool)
and returning int
```

### Example 1

```text
f(x,y)=x+y
```

Possible inferred type:

```text
(int,int) -> int
```

because `+` usually requires numeric operands.

---

### Example 2

```text
f(g,x)=g(x)
```

Possible type:

```text
((int -> bool), int) -> bool
```

because:

* `g` must be a function
* `x` is passed into `g`
* result type is whatever `g` returns

---

### Common JVM Style

```text
(II)I
```

means:

```text
(int,int) -> int
```

# 6. Alignment & Padding

### Common Questions Students Should Be Able to Answer

* Why does padding exist?
* How to compute structure size?
* Why can actual structure size exceed field sizes?
* How can field ordering reduce padding?

Memory alignment places data at addresses suitable for hardware. Padding may be inserted automatically.

### Example

**Question:**

Given the following C/C++-like declaration:

```c
union {
    struct {
        char alo;
        int detem;
        bool tay;
    } domixi;

    struct {
        bool bjj;
        double nhe_thoi_anh_oi;
        bool em_rot_mon_mat;
    } trinhhalan;
};
```

Assume:

```text
char = 1 byte
bool = 1 byte
int = 4 bytes
double = 8 bytes
```

Find:

1. Size of `domixi` without alignment
2. Size of `trinhhalan` without alignment
3. Size of the `union` without alignment
4. Size of `domixi` with alignment
5. Size of `trinhhalan` with alignment
6. Size of the `union` with alignment

---

## Answer

### Without alignment

`domixi`:

```text
char + int + bool
= 1 + 4 + 1
= 6 bytes
```

`trinhhalan`:

```text
bool + double + bool
= 1 + 8 + 1
= 10 bytes
```

Union size = largest member:

```text
max(6, 10) = 10 bytes
```

---

## With alignment

Assume each type aligns to its own size:

```text
char/bool → 1-byte aligned
int       → 4-byte aligned
double    → 8-byte aligned
```

### `domixi`

```c
struct {
    char alo;   // 1
    int detem;  // 4
    bool tay;   // 1
}
```

Layout:

```text
offset 0: char alo          → 1 byte
offset 1-3: padding         → 3 bytes
offset 4-7: int detem       → 4 bytes
offset 8: bool tay          → 1 byte
offset 9-11: padding        → 3 bytes
```

Why final padding?

```text
struct alignment = max member alignment = 4
size must be multiple of 4
9 → rounded to 12
```

So:

```text
domixi size = 12 bytes
padding = 3 + 3 = 6 bytes
```

### `trinhhalan`

```c
struct {
    bool bjj;                  // 1
    double nhe_thoi_anh_oi;    // 8
    bool em_rot_mon_mat;       // 1
}
```

Layout:

```text
offset 0: bool bjj                         → 1 byte
offset 1-7: padding                        → 7 bytes
offset 8-15: double nhe_thoi_anh_oi        → 8 bytes
offset 16: bool em_rot_mon_mat             → 1 byte
offset 17-23: padding                      → 7 bytes
```

Why final padding?

```text
struct alignment = max member alignment = 8
size must be multiple of 8
17 → rounded to 24
```

So:

```text
trinhhalan size = 24 bytes
padding = 7 + 7 = 14 bytes
```

### Union with alignment

A union size must:

1. Be large enough for the largest member
2. Be aligned to the strictest member alignment

Members:

```text
domixi       = 12 bytes, alignment 4
trinhhalan   = 24 bytes, alignment 8
```

Largest member = `trinhhalan`:

```text
24 bytes
```

Strictest alignment = 8.

Since:

```text
24 is already multiple of 8
```

Final:

```text
union size with alignment = 24 bytes
```

# 7. Stack Frames & Activation Records

<img src='https://preview.redd.it/data-structure-for-adults-v0-wey32dcahr261.jpg?auto=webp&s=958902b92f8ae686e32fd6cfcb53ec94730a3517' height=300 />

The **stack effect of a JVM instruction is fixed** by the JVM specification.
It does **NOT depend on the surrounding code**.

For example:

| Instruction | Always does            |
| ----------- | ---------------------- |
| `iadd`      | pop 2 ints, push 1 int |
| `if_icmpne` | pop 2 ints             |
| `dup`       | push 1 extra copy      |
| `iastore`   | pop 3 values           |

So:

```text id="z0jzai"
stack change = determined by the instruction itself
```

---

Example:

```text id="hwy0t2"
iadd
```

Before:

```text id="mcl3g4"
[..., a, b]
```

After:

```text id="t2rq8e"
[..., a+b]
```

Net effect:

```text id="1ajybg"
-1
```

because:

* pop 2
* push 1

---

Example:

```text id="1w2t8z"
iconst_1
iconst_2
iadd
```

Stack sizes:

```text id="0okbzj"
0 → 1 → 2 → 1
```

`iadd` always changes stack by `-1`, regardless of context.

### Common table

| Instruction | Stack effect | Reason                              |
| ----------- | -----------: | ----------------------------------- |
| `bipush`    |           +1 | pushes an int constant              |
| `baload`    |           -1 | pops arrayref + index, pushes value |
| `iadd`      |           -1 | pops 2 ints, pushes result          |
| `iconst_1`  |           +1 | pushes constant 1                   |
| `fastore`   |           -3 | pops arrayref + index + value       |
| `fstore_3`  |           -1 | pops one float into local slot 3    |
| `goto`      |            0 | jumps only                          |
| `if_icmpne` |           -2 | pops 2 ints to compare              |

---



# 8. Memory Management Problems

### Common Questions Students Should Be Able to Answer

* What is the difference between dangling reference and memory leak?
* How do memory leaks happen?
* Why are dangling references dangerous?
* What is aliasing?

### Common Problems

* Dangling reference
* Memory leak
* Aliasing

### Dangling Reference

Reference points to deallocated memory.

### Memory Leak

Allocated memory becomes unreachable.

### Example

```c
int *p = malloc(sizeof(int));
p = NULL;
```

# 9. Name Binding & Scope

### Scope Types

- Static Scope: Uses program structure.
- Dynamic Scope: Uses calling sequence.

### Binding
- Deep binding
- Shallow binding

### Example

**Question:**

Consider the following pseudo-code with **dynamic scope**:

```python id="89w4nh"
x = 'MiuLe'

def maithuy():
    print(x)

def TuNgay(f):
    x = 'HaiDangDoo'
    maithuy()

TuNgay(maithuy)
```

What will be printed under:

1. Deep binding
2. Shallow binding

---

### Answer

#### Deep binding: `MiuLe`

Reason:
Use environment when function was passed/defined

> `maithuy` remembers: `x = 'MiuLe'`

#### Shallow binding: `HaiDangDoo`

Reason: Use environment at call time

> When `f()` executes inside `TuNgay`, current `x` is `'HaiDangDoo'`

---

### Example 2

```text id="u8ewh7"
Binding strategy decides WHICH environment to use,
but global/nonlocal can override where variables are searched.
```

**Question:**

Consider the following pseudo-code with **dynamic scope**:

```python id="vx8r8z"
x = 'MiuLe'

def maithuy():
    global x  # Line 4
    print(x)

def TuNgay(f):
    x = 'HaiDangDoo'

    def yeuMotNguoiCoLe():
        nonlocal x
        x = 'LouHoang'

    yeuMotNguoiCoLe()
    maithuy()

TuNgay(maithuy)
```

What will be printed under:

1. Deep binding
2. Shallow binding
3. Deep binding, if Line 4 is removed
4. Shallow binding, if Line 4 is removed
---

### Case 1,2: `MiuLe`

Reason:

> global x forces lookup at global scope, regardless of binding

⇒ `maithuy()` always uses: x = 'MiuLe'

Note: `nonlocal` only changes the `x` inside `TuNgay`

### Case 3: `MiuLe`
Same reason as previous question.

### Case 4: `LouHoang`
> `nonlocal` accessed `x` in `TuNgay`, and changed its value to `LouHoang`


| Keyword    | Meaning                          |
| ---------- | -------------------------------- |
| `local`    | current function scope           |
| `global`   | global/module scope              |
| `nonlocal` | nearest enclosing function scope |




---

# 10. Parameter Passing Methods

### Common Questions Students Should Be Able to Answer

* When would you prefer pass-by-value over pass-by-reference?
* Which methods can modify caller variables?
* Which methods may cause aliasing?
* Why is pass-by-reference more efficient for large objects?

### Main Methods

| Method               | Behavior               |
| -------------------- | ---------------------- |
| Pass by value        | copy                   |
| Pass by reference    | alias                  |
| Pass by value-result | copy-in/out            |
| Pass by name         | re-evaluate expression |

### Example

```cpp
void f(int &x) {
    x = 10;
}
```

# 15. Array Ordering

**Question:**

Given the declaration:

```c id="5q8jjr"
int A[4][5][6];
```

Assume:

* `int` occupies 4 bytes
* Base address of `A[0][0][0]` is 0

Find the address of:

```c id="w1xq5m"
A[2][3][4]
```

for BOTH:

1. Row-major order
2. Column-major order

---

## Answer

### 1. Row-major order

Formula:

```text id="5rkdz9"
((i × d2 × d3) + (j × d3) + k) × size
```

Substitute:

```text id="g9tm8n"
((2 × 5 × 6) + (3 × 6) + 4) × 4
= (60 + 18 + 4) × 4
= 82 × 4
= 328
```

Answer:

```text id="q8rqpa"
328
```

---

### 2. Column-major order

Formula:

```text id="rf8cd3"
((k × d1 × d2) + (j × d1) + i) × size
```

Substitute:

```text id="zg9zpc"
((4 × 4 × 5) + (3 × 4) + 2) × 4
= (80 + 12 + 2) × 4
= 94 × 4
= 376
```

Answer:

```text id="fbj02q"
376
```
